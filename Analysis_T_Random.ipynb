```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
x_col=['DegC','SDg']
y_col='Tmax'
```

```python
X=df[x_col]
y=df[y_col]
```

```python
test_size=.2
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=test_size, 
                                                    random_state=1)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

pm = LinearRegression()
pm.fit(X_train_poly, y_train)
```

```python
pm.intercept_, pm.coef_
```

```python
pm.score(X_train_poly, y_train) #R-squared value
```

```python
pm.intercept_ + sum((pm.coef_ * X_test_poly[0]))
```

```python
z=pm.predict(X_test_poly)
z
```

```python
fig=plt.figure(figsize=(8,6))
plt.scatter(w, z, label='Predicted Value')
plt.scatter(w, y_test, label='Real Value')
plt.xlabel("$_{EFF}$ (1/year)")
plt.ylabel("$T_{max}$ (Year)")
#plt.title("Tmax vs Effective first order decay coefficient")
plt.legend()
fig.savefig('predictionD.jpg', bbox_inches='tight', dpi=150)
plt.show()
```

```python
fig=plt.figure(figsize=(8,6))
plt.scatter(v, z, label='Predicted Value')
plt.scatter(v, y_test, label='Real Value')
plt.xlabel("$$ (1/year)")
plt.ylabel("$T_{max}$ (Year)")
#plt.title("Tmax vs Source Decay Coefficient")
plt.legend()
fig.savefig('PredictionS.jpg', bbox_inches='tight', dpi=150)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, z)
print(mse)

```
